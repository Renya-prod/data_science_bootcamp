```python
import pandas as pd
import sqlite3

db_path = "checking-logs.sqlite"
conn = sqlite3.connect(db_path)

tables_query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql(tables_query, conn)

print(tables)
```

#### 4. SQL-–∑–∞–ø—Ä–æ—Å –¥–ª—è –≤—ã–±–æ—Ä–∫–∏ –¥–∞–Ω–Ω—ã—Ö —Å —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–µ–π –∏ —á—Ç–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –≤ DataFrame

```python
query = """
SELECT uid, datetime
FROM pageviews
WHERE uid LIKE 'user_%'
ORDER BY uid ASC;
"""

pageviews = pd.io.sql.read_sql(query, conn)

print("\n–û—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω—ã–µ –∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –∏–∑ —Ç–∞–±–ª–∏—Ü—ã pageviews:")
print(pageviews.head(5))
```

#### 5. –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ datetime –≤ —Ñ–æ—Ä–º–∞—Ç DatetimeIndex

```python
pageviews["datetime"] = pd.to_datetime(pageviews["datetime"])
pageviews.set_index("datetime", inplace=True)

print(pageviews.head(5))
```
